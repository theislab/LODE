```python
OCT_PROJ_DIR = "/home/icb/olle.holmberg/projects/OCT_classification"
map_csv = os.path.join(OCT_PROJ_DIR, "data/data_visus_all_extended.csv")

map_pd = pd.read_csv(map_csv)


# pre process month, day
map_pd["MONTH"] = map_pd.MONTH.astype(str).str.zfill(2)
map_pd["DAY"] = map_pd.DAY.astype(str).str.zfill(2)

# create study date column
map_pd["STUDYDATE"] = map_pd.YEAR.astype(str) + map_pd.MONTH + map_pd.DAY
```

```python
#### merge
```

```python
merged = pd.merge(f, map_pd[["FILENAME", "SERIESID", "STUDYDATE", "LOK", "PATNR"]], 
         left_on ="filename", right_on="FILENAME", how="left")
```

```python
merged = merged.rename(columns={"0":"bg","1":"em","2":"nr","3":"if","4":"sf","5":"srhm",
                                "6":"rpe","7":"fp","8":"drusen",
                                "9":"pm","10":"choroid",
                                "11":"sp","12":"af","13":"fibrosis","14":"vitreous","15":"camera_effect"})
```
