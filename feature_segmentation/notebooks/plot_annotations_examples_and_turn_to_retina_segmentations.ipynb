```python
import os
import numpy as np
import matplotlib.pyplot as plt
import PIL.Image
import matplotlib.image as mpimg
import pandas as pd
from PIL import Image
import cv2

path = "/home/olle/PycharmProjects/longitudinal_eyeclinic_project/\
clinical_features_samples/data_dataset_voc/SegmentationClass"
save_annotation_path = "/home/olle/PycharmProjects/thickness_map_prediction/\
retinal_thickness_segmentation/data/clinic_data/long_anno_lbl"
label_paths = [os.path.join(path,i) for i in os.listdir(path)]

def change_to_retina_annotation(image):
    image[np.where(image>0)]=1
    for i in range(0,image.shape[1]):
        try:
            min_idy = np.min(np.where(image[:,i]>0))
            max_idy = np.max(np.where(image[:,i]>0))
            image[min_idy:max_idy,i] = 1
        except:
            pass
            #print("ValueError: zero-size array to reduction operation minimum which has no identity")
    return(image)
```

```python
from matplotlib.pyplot import figure
figure(figsize=(20, 220))

for label_path in label_paths:
    record_name = label_path.split("/")[-1].replace(".jpg",".jpeg")
    image = np.array(Image.open(label_path)).astype(np.float32)
    #image = cv2.imread(label_path)
    retina_image = change_to_retina_annotation(image)
    retina_image[retina_image == 2] = 0
    cv2.imwrite(os.path.join(save_annotation_path,record_name),retina_image)
    #plt.subplot(k, 3, i)
    #plt.imshow(retina_image)
    #plt.title("segmentation label")
    #plt.axis("off")
    
    #plt.savefig("./thiagos_annotations")
```
