```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data_file = "/home/olle/PycharmProjects/LODE/workspace/sequence_data/longitudinal_properties.csv"
naive_data_file = "/home/olle/PycharmProjects/LODE/workspace/sequence_data/check_naive_patients_CORRECTED.csv"

data_pd = pd.read_csv(data_file)
naive_data_pd = pd.read_csv(naive_data_file, sep=",|;")

data_pd.head()
```

```python
naive_data_pd
```

#### format naive pd

```python
naive_data_pd = naive_data_pd[["patient_id", "laterality", "first_injection_date", "Naive"]]
naive_data_pd["sequence"] = naive_data_pd["patient_id"].astype(str) + "_" + naive_data_pd["laterality"]
```

```python
naive_data_pd
```

```python
print(f"Number of sequences in data pd {data_pd.shape[0]} number of in naive pd {naive_data_pd.shape[0]}")
```

#### merge tables

```python
sequence_naive_pd = pd.merge(naive_data_pd, data_pd, left_on="sequence", right_on="sequence", how="left")
```

```python
sequence_naive_pd = sequence_naive_pd.dropna(subset=["Naive"])

# filter for naive records 
naive_bool = (sequence_naive_pd.Naive == 1) | (sequence_naive_pd.Naive == 4)

print("number of naive sequences are: ", sum(naive_bool))
```

```python
#### save naive longitudinal properties table
```

```python
save_path = "/home/olle/PycharmProjects/LODE/workspace/sequence_data/longitudinal_properties_naive.csv"
sequence_naive_pd.loc[naive_bool].to_csv(save_path)
```
