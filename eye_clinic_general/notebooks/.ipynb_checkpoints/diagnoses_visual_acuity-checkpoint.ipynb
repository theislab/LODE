```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#pd.set_option('display.max_rows', 500)
#pd.set_option('display.height', 1000)

dwh_path = "/home/olle/PycharmProjects/eye_clinic_general/dwh_information"
data_export_path = "./oct_files.csv"
dkey_codes_path = "2018-ICD-10-Code-Descriptions"

# dkey lookup table
dkey_codes_pd = pd.read_fwf(os.path.join(dwh_path, dkey_codes_path,"icd10cm_codes_2018.txt"), 
                            sep=" ", header=None)

# diagnosis and oct table
diagnoses_pd = pd.read_csv(os.path.join(dwh_path,"diagnosis.csv"),index_col=False).drop("Unnamed: 0",axis=1)
oct_file_pd = pd.read_csv(os.path.join(dwh_path,data_export_path)).drop("Unnamed: 0",axis=1)
```

### Process the diagnosis key table

```python
#extract main title of description
dkey_codes_pd[1] = dkey_codes_pd[1].str.split(",",expand=True)[0]

#remove all to granular code defenitions
dkey_codes_pd[0] = dkey_codes_pd[0].str[0:4]
dkey_codes_filt_pd = dkey_codes_pd[dkey_codes_pd[0].str.len() <= 4]

#create the DWH DKEY format for joining 
first_key = dkey_codes_filt_pd[0].str[0:3]
scnd_key = dkey_codes_filt_pd[0].str[3:4]
#create new column for joining
dkey_codes_filt_pd["DKEY"] = first_key + "." + scnd_key

#rename descr column
dkey_codes_filt_pd = dkey_codes_filt_pd.rename(columns={1:"desc"}).drop_duplicates([0,"DKEY"])
```

### Process the data export

```python
#split path into columns
oct_pd = oct_file_pd["0"].str.split("/",expand=True)[[1,2,3]]

#rename columns
oct_pd = oct_pd.rename(columns={1:"patient_id",2:"laterality",3:"study_date"})

#remove unknown records 
oct_pd = oct_pd[oct_pd.patient_id != "Unkown"]

#change laterality encoding
oct_pd.laterality =  oct_pd.laterality.str.replace("Right","R")
oct_pd.laterality =  oct_pd.laterality.str.replace("Left","L")

#set dtypes
oct_pd.patient_id = oct_pd.patient_id.astype(int)
```

#### Split B laterality to left and right

```python
diagnoses_pd_LR = diagnoses_pd[diagnoses_pd.LOK != "B"]

both_lats_R = diagnoses_pd[diagnoses_pd.LOK == "B"]
both_lats_L = diagnoses_pd[diagnoses_pd.LOK == "B"]

both_lats_R.LOK = "R"
both_lats_L.LOK = "L"

#join both L and R rows
diagnosis_lr_pd = pd.concat([both_lats_R,both_lats_L,diagnoses_pd_LR])

#reformat date columns
diagnosis_lr_pd.DAT = diagnosis_lr_pd.DAT.str.replace("-","")

#remove all digits after first in diagnose name
dkey = diagnosis_lr_pd.DKEY.str.split(".",expand=True)
first_digit = dkey[1].str[0:1]
diagnosis_lr_pd["DKEY"] = dkey[0] + "." + first_digit
```

```python
print("Number of oct reocrds: {}, Number of diagnosis records: {}".format(oct_file_pd.shape[0], 
                                                                         diagnosis_lr_pd.shape[0]))
```

### Create look up table based on patient id and laterality

```python
#patient_lookup
diag_pat_lookup = diagnosis_lr_pd[["PATNR"]].drop_duplicates().dropna()
oct_pat_lookup = oct_pd[["patient_id"]].drop_duplicates().dropna()

oct_and_diag_pat_lookup = pd.merge(oct_pat_lookup[["patient_id"]], diag_pat_lookup[["PATNR"]],
         left_on ="patient_id", right_on="PATNR", how="inner").drop_duplicates()

#patient and laterality lookup 
diag_pat_lat_lookup = diagnosis_lr_pd[["PATNR","LOK"]].drop_duplicates().dropna()
oct_pat_lat_lookup = oct_pd[["patient_id","laterality"]].drop_duplicates().dropna()

oct_and_diag_pat_lat_lookup = pd.merge(oct_pat_lat_lookup[["patient_id","laterality"]], 
                                       diag_pat_lat_lookup[["PATNR","LOK"]],
         left_on =["patient_id","laterality"], right_on=["PATNR","LOK"], 
                                       how="inner").drop_duplicates()

#patient, laterality and study date lookup
oct_pat_lat_date_lookup = oct_pd[["patient_id","laterality","study_date"]].drop_duplicates().dropna()

oct_and_diag_pat_lat_dat_lookup = pd.merge(oct_pat_lat_date_lookup[["patient_id","laterality","study_date"]], 
                                       diag_pat_lat_lookup[["PATNR","LOK"]],
         left_on =["patient_id","laterality"], right_on=["PATNR","LOK"], 
                                       how="inner").drop_duplicates()
```

#### Account for the number fall out of octs

```python
print("Number of patients with diagnosis: {},witth oct {},\
with oct and some diagnosis: {}".format(diag_pat_lookup.shape[0],
                                                        oct_pat_lookup.shape[0],
                                                          oct_and_diag_pat_lookup.shape[0]))
```

```python
print("Number of patients and eyes with diagnosis: {}, with oct {},\
with oct and some diagnosis: {}".format(diag_pat_lat_lookup.shape[0],
                                                        oct_pat_lat_lookup.shape[0],
                                                          oct_and_diag_pat_lat_lookup.shape[0]))
```

```python
print("Number of patients, eyes and dates with with oct {},\
with oct and some diagnosis: {}".format(oct_pat_lat_date_lookup.shape[0],
                                                        oct_and_diag_pat_lat_dat_lookup.shape[0]))
```

#### view patients/eyes where no diagnosis is available

```python
oct_exclusion = pd.merge(oct_pat_lat_lookup[["patient_id","laterality"]], 
                                       diag_pat_lat_lookup[["PATNR","LOK"]],
         left_on =["patient_id","laterality"], right_on=["PATNR","LOK"], 
                                       how="outer",indicator=True).drop_duplicates()

#filter for all excluded records
oct_exclusion_ = oct_exclusion[oct_exclusion["_merge"]=="left_only"]
#make key to int
oct_exclusion_.patient_id = oct_exclusion_.patient_id.astype(int)
oct_exclusion_.head()
```

### Merge the diagnosis and oct tables based on id and laterality

```python
#rename columns for clarity
diagnosis_abt_pd = diagnosis_lr_pd.rename(columns={"DAT":"date_diagnosis","LOK":"dkey_laterality"})
#drop dkat info
diagnosis_abt_pd = diagnosis_abt_pd.drop("DKAT",axis=1)
```

```python
oct_and_diag_pd = pd.merge(oct_pd,diagnosis_abt_pd,
         left_on =["patient_id","laterality"], right_on=["PATNR","dkey_laterality"], 
                                       how="inner").drop_duplicates()
#remove all rows created with dkey given before oct obtained 
diagnosis_before_oct_ = oct_and_diag_pd.date_diagnosis < oct_and_diag_pd.study_date

#drop duplicates w.r.t. patient, eye, study date and diagnosis
oct_and_diag_abt = oct_and_diag_pd[~diagnosis_before_oct_].drop_duplicates(["patient_id",
                                                         "laterality",
                                                         "DKEY",
                                                         "study_date"]).dropna()

print("The number of oct's with determined diagnosis are: {}".format(oct_and_diag_abt.shape[0]))
oct_and_diag_abt.head(10)
```

### all patients with diabetic diagnosises

```python
#filter out all E diagnises, i.e. with diabetes
oct_diag_diab_abt = oct_and_diag_abt[oct_and_diag_abt.DKEY.str.contains('E',case=False)]

#combine id strings to ids to create unique record ids
oct_diag_diab_abt["id"] = oct_diag_diab_abt["patient_id"].astype(str)\
+"_"+oct_diag_diab_abt["laterality"].astype(str)\
+"_"+oct_diag_diab_abt["study_date"].astype(str)
```

# Analyze longitudinal records

```python
#get boolean for patients with over 

number_of_studies_threshold = 10
over_five_studies = oct_and_diag_abt.groupby('patient_id')['study_date'].nunique() > number_of_studies_threshold

#get patient ids with over 5 studies made with OCT's
patients_over_five_studies = pd.DataFrame(over_five_studies).reset_index()
patients_over_five_studies = patients_over_five_studies[patients_over_five_studies.study_date\
                                                        ==True][["patient_id"]]

#create complete and date sorted abt
longitudinal_abt = pd.merge(patients_over_five_studies, oct_and_diag_abt, 
         left_on="patient_id",right_on="patient_id",how="inner").sort_values(['patient_id','study_date'])

#get number of patients available
num_patients = np.unique(longitudinal_abt.patient_id).shape[0]

print("The number of patient with more then {} studies made are: {}".format(number_of_studies_threshold,
                                                                            num_patients))
```

# Visual acuity 

```python
import re
from fractions import Fraction
import pandas as pd
pd.set_option('display.max_rows', 500)
pd.set_option('display.height', 1000)

dwh_path = "/home/olle/PycharmProjects/eye_clinic_general/dwh_information"
data_export_path = "./oct_files.csv"

va_pd = pd.read_csv(os.path.join(dwh_path,"visus_labels.csv"),index_col=False).drop("Unnamed: 0",axis=1)
oct_file_pd = pd.read_csv(os.path.join(dwh_path,data_export_path)).drop("Unnamed: 0",axis=1)
```

### Process the visual acuity table

```python
#drop missing values
va_pd_ = va_pd.dropna(subset=['visual_acuity_VISUS'])
#drop values conaining no digit
filter_vector = va_pd_.visual_acuity_VISUS.str.contains('\d', regex=True)
va_pd_ = va_pd_[filter_vector]

va_pd_.visual_acuity_VISUS = va_pd_.visual_acuity_VISUS.str.replace(r'\W+', '.')

va_pd_.MEASUREMENT_DATE = va_pd_.MEASUREMENT_DATE.str.replace("-","")
```

### join with OCT table

```python
joint_pd = pd.merge(va_pd_,oct_pd,left_on=["PATNR","AUGE","MEASUREMENT_DATE"],
                    right_on=["patient_id","laterality",
                    "study_date"],how="inner").drop_duplicates()
```

#### Convert visus column to numeric

```python
joint_pd.visual_acuity_VISUS = joint_pd.visual_acuity_VISUS.convert_objects(convert_numeric=True)
```

```python
joint_pd = joint_pd.dropna(subset=["visual_acuity_VISUS"])
```

```python
va_abt = pd.DataFrame(joint_pd.groupby(['PATNR','AUGE','study_date'], sort=False)['visual_acuity_VISUS'].max())
va_abt = va_abt.reset_index()
```

```python
pd.cut(va_abt.visual_acuity_VISUS[va_abt.visual_acuity_VISUS<1.5],14).value_counts()
```
